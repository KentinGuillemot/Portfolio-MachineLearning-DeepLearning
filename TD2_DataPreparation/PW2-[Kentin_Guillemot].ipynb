### Gestion valeurs manquantes par colonnes 

```python
def missing_colums(originDB):
  colsWithMissing = []

  for col in originDB.columns:
    if originDB[col].isnull().any():
      colsWithMissing.append(col)

  reduced_original_data = originDB.drop(colsWithMissing, axis=1)

  return colsWithMissing, reduced_original_data
```

```python
numerical_cols = melb_data.select_dtypes(exclude=['object']).columns
categorical_cols = melb_data.select_dtypes(exclude=['number']).columns

print("Numerical columns:", numerical_cols)
print("Categorical columns:", categorical_cols)
```

```python
for col in categorical_cols:
    print(col + ":", melb_data[col].nunique())
```

Etant donner que je vais faire du one-hot encoding dans la pipeline je vais devoir garder seulement les colonnes avec un nunique < 10 car sinon cela ne sera plus pertinent

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier


# Trouver les colonnes catégorielles avec moins de 10 catégories pour le OneHotEncoder (car sinon au dessus ce n'est pas optimal)
onehot_columns = [col for col in categorical_cols if melb_data[col].nunique() < 10]

# Trouver les autres colonnes catégorielles pour OrdinalEncoder
ordinal_columns = [col for col in categorical_cols if col not in onehot_columns]

# Pipeline pour les colonnes numériques (ex: imputation)
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')), #remplace les valeurs manquantes par la median (souvent meilleur methode)
    ('scaler', StandardScaler()) #centre et reduit les données
])

# Pipeline pour les colonnes OneHot
onehot_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),  #remplace les valeurs manquantes par la valeur la plus fréquente
    ('onehot', OneHotEncoder(handle_unknown='ignore')) #crée des colonnes pour chaque catégorie et met 1 si la catégorie est présente et 0 sinon
])

# Pipeline pour les colonnes Ordinal
ordinal_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')), #remplace les valeurs manquantes par la valeur la plus fréquente
    ('ordinal', OrdinalEncoder()) #transforme les valeurs catégorielles en valeurs numériques
])

# Combiner le tout dans un ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_cols),
        ('onehot', onehot_transformer, onehot_columns),
        ('ordinal', ordinal_transformer, ordinal_columns)
    ]
)

# Pipeline final avec modèle de classification (exemple RandomForest)
model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('pca', PCA(n_components=4)), #réduit la dimensionnalité du dataset
    ('classifier', RandomForestClassifier())
])


"""

Ce n'est pas demandé mais par la suite il faudrait faire le train test split et fit le modèle

"""
```
