## Saving data in different formats

Information on a computer is stored in files. Each file, with its unique name, a period, and an extension, plays a crucial role in our digital world. The file name is not just a label but a key that helps us differentiate one file from another. Conversely, the extension guides us on how the data is stored in the file **(.csv comma-separated values**) and, indirectly, the application that allows the file to be opened **(.doc is a Microsoft Word document)**. The extension, typically three letters, with some exceptions (.jpeg, .html, .json, etc.), is a vital part of the file's identity.

Some file extensions are characteristic when it comes to storing data sets. Files represented by these extensions can be grouped into two large groups: flat files and rich files. The first file type stores alphanumeric characters (e.g., txt, CSV, JSON, etc.). Rich files allow, in addition to flat data, to assign other characteristics such as metadata, font size, and color (e.g., doc, xls, etc.). In this TP, we will concentrate only on flat files.

Pandas allows us to save *dataframes* in different formats (i.e., CSV, TSV, json). For example, if we want our *dataframe*, which contains clean and discretized data, to be stored in a *.csv* (*comma separated values*) file, we must add the following line:

```python
#from google.colab import files # Please disable this comment if Google Collab is used.
df.to_csv("output.csv", sep=',', encoding='utf-8')
#files.download("output.csv") # Please disable this comment if Google Collab is used.
```

The line above saves the data from the *df* (*data frame*) in a file called *output.csv*. This file uses the comma (,) to separate the attributes in the utf-8 encoding shape (this last parameter can be omitted). The output file is saved in the same directory where our Python script (notebook) is located. To see its content, we can run the command *cat output.csv* in the Linux command line (use *type* in the Windows operating system).

If we want to store a *tsv* file (tab-separated values), we change the previous line to the following.

```python
df.to_csv("output.tsv", sep='\t', encoding='utf-8')
```

We can also transform the data into more complex structures like a JSON file.

```python
df.to_json("output.json", orient = 'columns')
```

# Dataset car.csv :

```python
#from google.colab import files # Please disable this comment if Google Collab is used.
#uploaded = files.upload() # Please disable this comment if Google Collab is used.
data = 'cars.csv'
car = pd.read_csv(data, sep=";", engine='python')
```

```python
car.head()
```

```python
car.drop(0, inplace=True) #Je retire la première ligne car inutile (c'est juste le type des datas)
```

```python
car
```

```python
car.dtypes #je regarde le type des datas
```

```python
#la fonction astype() sert à convertir les datas d'une colonnes dans un type en particulier
car["Car"] = car["Car"].astype(str)
car["MPG"] = car["MPG"].astype(float)
car["Cylinders"] = car["Cylinders"].astype(int)
car["Displacement"] = car["Displacement"].astype(float)
car["Horsepower"] = car["Horsepower"].astype(float)
car["Weight"] = car["Weight"].astype(float)
car["Acceleration"] = car["Acceleration"].astype(float)
car["Model"] = car["Model"].astype(int)
car["Origin"] = car["Origin"].astype(str)
```

```python
car.dtypes #je regarde si les modifications ont marché
```

```python
car.isnull().sum() #je regarde si y a des NAN values
```

```python
car.info() #sert à avoir plusieurs info en meme temps
```

```python
car.describe() #sert à avoir des information mathématique comme min, max, mean ...
```

```python
car.columns
```

```python
car[["Car","Acceleration"]][car["Acceleration"]>23].sort_values(by="Acceleration", ascending=False) #Donne les 4 voitures ayant une accelaration > 23 et dans l'ordre décroissant
```

```python
car[["Car","Horsepower"]].sort_values(by="Horsepower", ascending=False).head() #Nous donne les 5 voitures avec le plus de Horsepower DANS L'ORDRE DECROISSANT
```

```python
car["Cylinders"].value_counts() #Donne le nombre d'element possédant chaque value
```

```python
#j'ai créé une nouvelle colonne calculant le rapport poid puissance servant à indiqué si une voiture est puissante ou non
#plus le rapport est bas plus la voiture est puissante

car["Poid/Puissance"] = car["Weight"]/car["Horsepower"]
car.head()
```

```python
car[["Car","Poid/Puissance"]].sort_values(by="Poid/Puissance").head() #Voici les 5 voitures les plus puissantes
```
