# Exercice 2

### 1)

```python
from sklearn.pipeline import Pipeline

def LR_Pipeline(scaler):
    
    my_pipeline = Pipeline([
        ('scaler', scaler),
        ('model', LinearRegression())
    ])
    
    return my_pipeline
```

### 2)

```python
def LR_evalutation(X, y, scaler, test_size=0.2):
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=20, stratify=y)
    
    model = LR_Pipeline(scaler)
    
    model.fit(X_train, y_train)
    
    predictions = model.predict(X_test)
    
    print('Mean squared error: %.2f' % mean_squared_error(y_test, predictions))
    print('Coefficient of determination: %.2f' % r2_score(y_test, predictions))
    
    plt.scatter(y_test, predictions)
    plt.xlabel('True Values')
    plt.ylabel('Predictions')
    plt.title('Predictions vs True Values')
    plt.show()
```

### 3)

```python
LR_evalutation(X, y, StandardScaler())
```

### 2)

```python
def Reg_Pipeline(scaler, model):
    pipeline = Pipeline([
    ('scaler', scaler),    
    ('model', model) 
    ])
    return pipeline
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

accuracy = accuracy_score(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print("Confusion Matrix:\n", confusion)
print("Classification Report:\n", classification_rep)
```

The recall is detecting the quantity of True positive that preddict the model in all the real True positive. Here the model has a really bad recall for the "0" class. 

The f1-score is basically the mean between de precision and the recall, here due to the bad recall score we have a bad f1-score for the "0" class

```python
LogReg=LogisticRegression()
naive=GaussianNB()
scaler=StandardScaler()
```

```python
mypipe_logreg=Reg_Pipeline(scaler,LogReg)
mypipe_naive=Reg_Pipeline(scaler,naive)
```

```python
y_prob_naive=mypipe_naive.predict_proba(X_test)[:, 1]
y_prob_logreg=mypipe_logreg.predict_proba(X_test)[:, 1]
#we predict by probability for the auc/roc
```

```python
# Now we gonna evaluate the models

auc_lr = roc_auc_score(y_test, y_prob_logreg)
auc_nb = roc_auc_score(y_test, y_prob_naive)

print(f"Logistic Regression AUC: {auc_lr:.2f}")
print(f"Naive Bayes AUC: {auc_nb:.2f}")

y_pred_lr = mypipe_logreg.predict(X_test)
y_pred_nb = mypipe_naive.predict(X_test)

print("Logistic Regression Classification Report:")
print(classification_report(y_test, y_pred_lr))
print("Confusion matrix Logreg :")
print(confusion_matrix(y_test,y_pred_lr))

print("Naive Bayes Classification Report:")
print(classification_report(y_test, y_pred_nb))
print("Confusion matrix Naive :")
print(confusion_matrix(y_test,y_pred_nb))
```

```python
fpr_lr, tpr_lr, _ = roc_curve(y_test, y_prob_logreg)
fpr_nb, tpr_nb, _ = roc_curve(y_test, y_prob_naive)

plt.figure(figsize=(8, 6))
plt.plot(fpr_lr, tpr_lr, label=f'Logistic Regression (AUC = {auc_lr})')
plt.plot(fpr_nb, tpr_nb, label=f'Naive Bayes (AUC = {auc_nb})')
plt.plot([0, 1], [0, 1], 'k--') 
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve Comparison')
plt.legend(loc="lower right")
plt.show()
```
